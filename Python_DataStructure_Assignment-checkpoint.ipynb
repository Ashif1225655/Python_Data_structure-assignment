```python
# (5). Describe the key features of sets and provide examples of their use.

# sets are the fundamental data structure that is used to store different type of data it is mutable, no indexing , 
# Hetrogeneous, and unordered, duplicate element not allowed.

# Mutable >> Mutable that means once created after that they can modifined also and manipulate 
# e.g 
my_set = {1,2,3,6,5,4}
my_set.add(7) # add() method is used to add the in last of sets
print(my_set) # output {1, 2, 3, 4, 5, 6, 7}

# Hetrogeneous >> that means the sets can holds the all type of data type.
hetro = {1,2.5,'Hello'}
print(hetro)

# remove() >> it is used to remove the element form sets

remv = {1,2,3,6,5}
remv.remove(3)
print(remv)


# Unique Elements >> The sets are the only contain the unique elements the there is not allow to duplicate elements in sets.
# e.g:
set = {1,2,3,3,4,5,4}
print(set)  # output {1, 2, 3, 4, 5}

# sets support the various mathematic operations such as union , intersection, difference and symmetric difference.
# e.g: 
# union() 0r | >> union is used to combine the unique element do not combine dupllicate refers only onet times

set_1 = {1,2,3,4,5}
set_2 = {4,5,6,7,8}

print(set_1 | set_2) # output {1, 2, 3, 4, 5, 6, 7, 8}

# intersection() or & >> it gives the only common elements in the both sets.
set_1 = {1,2,3,4,5}
set_2 = {4,5,6,7,8}

print(set_1 & set_2) # output {4, 5}

# Difference() or (-) >> it find the element that are exit in first set not in second set

set_1 = {1,2,3,4,5}
set_2 = {4,5,6,7,8}

print(set_1 - set_2) # output {1, 2, 3}

# symmetric_difference() or ^ >> except the common element provide all elements.

#e.g:

set_1 = {1,2,3,4,5}
set_2 = {4,5,6,7,8}

print(set_1 ^ set_2) # output {1, 2, 3}

# sets support the frozenset that make the sets immutable once created element that connot add or remove the 
# this when we want the sets is immutable

frozen = frozenset([1,2,3,4,5])
print(type(frozen))
```
